```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-5,5,0.25)
y = np.arange(-5, 5,0.25)
z = np.array(range(-5,6,1))
print(x, '\n\n', y, '\n\n', z)
x, y = np.meshgrid(x,y)
r = np.sqrt(x**2+y**2)
z = np.sin(r)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)
plt.draw()
plt.savefig('physics_modeling.jpg', format='jpg')
```

```python
from matplotlib.animation import FuncAnimation

x = []
y = []

figure, ax = plt.subplots()

ax.set_xlim(0, 100)
ax.set_ylim(0, 12)

line, = ax.plot(0, 0)

def anium(i):
    x.append(i * 15)
    y.append(i)
    
    line.set_xdata(x)
    line.set_ydata(y)
    return line,

animation = FuncAnimation(figure, func = anium, frames = np.arange(0, 10, 0.1), interval = 10)
plt.show()
```
