```python
# Body Mass Index (BMI) is a measurement of body fat,
# that indicate person is at a healthy weight.

# BMI calculator of five people,
# visualise their data with matplotlib.

# NOTE: 100cm = 1m
```

```python
bmis = []
names = []

def calc_bmi(name, height: "cm", weight: "kg"):

    convert_height = height / 100
    bmi = weight / (convert_height ** 2)
    
    if bmi < 18.5:
        check = "under-weight"
    elif bmi > 18.5 and bmi <= 24.9:
        check = "normal-weight"
    elif bmi > 25 and bmi <= 29.9:
        check = "over-weight"
    else:
        check = "obese"
        
    bmis.append(weight)
    names.append(name)
    
    return f"{name} BMI is : {bmi} ({check})"
```

```python
Usman = calc_bmi("Usman", 175, 55)
print(Usman)
```

```python
Benjamin = calc_bmi("Benjamin", 160, 70)
print(Benjamin)
```

```python
Kamala = calc_bmi("Kamala Haris", 157, 55)
print(Kamala)
```

```python
Donald = calc_bmi("Donald Trump", 190, 98)
print(Donald)
```

```python
Isaac = calc_bmi("Isaac Newton", 1.85, 85)
print(Isaac)
```

```python
import numpy as np
import matplotlib.pyplot as plt

xpoints = np.array(bmis)
ypoints = np.array(names)

plt.title("Visualise of five people BMI")
plt.ylabel("Names", color="green")
plt.xlabel("BMI(s)", color="red")

plt.bar(xpoints, ypoints)
plt.grid()
```
