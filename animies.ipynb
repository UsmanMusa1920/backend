```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots()

# Set background color
ax.set_facecolor('skyblue')

# Draw the ground
ax.add_patch(patches.Rectangle((0, 0), 1, 0.2, facecolor="green"))

# Draw the road
ax.add_patch(patches.Rectangle((0, 0.1), 1, 0.05, facecolor="gray"))

# Draw the car (red rectangle)
ax.add_patch(patches.Rectangle((0.35, 0.12), 0.2, 0.05, facecolor="red"))
# Car wheels (black circles)
ax.add_patch(patches.Circle((0.4, 0.1), 0.03, facecolor="black"))
ax.add_patch(patches.Circle((0.55, 0.1), 0.03, facecolor="black"))

# Draw the sun (yellow circle)
ax.add_patch(patches.Circle((0.85, 0.8), 0.1, facecolor="yellow"))

# Draw the hills (green triangles)
hill_coords1 = [(0.1, 0.2), (0.3, 0.5), (0.5, 0.2)]
hill_coords2 = [(0.4, 0.2), (0.7, 0.6), (0.9, 0.2)]
ax.add_patch(patches.Polygon(hill_coords1, closed=True, facecolor='green'))
ax.add_patch(patches.Polygon(hill_coords2, closed=True, facecolor='green'))

# Function to draw a windmill
def draw_windmill(x, y):
    # Tower (white)
    ax.plot([x, x], [y, y + 0.25], color='white', linewidth=5)
    # Blades (pink lines)
    ax.plot([x, x + 0.1], [y + 0.25, y + 0.35], color='pink', linewidth=3)
    ax.plot([x, x - 0.1], [y + 0.25, y + 0.35], color='pink', linewidth=3)
    ax.plot([x, x], [y + 0.25, y + 0.35], color='pink', linewidth=3)

# Draw three windmills
draw_windmill(0.2, 0.25)
draw_windmill(0.5, 0.25)
draw_windmill(0.8, 0.25)

# Set limits and hide axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Display the image
plt.show()
```
