```python
import numpy as np
import matplotlib.pyplot as plt

countries = [
    "India", "China", "United States", "Indonesia", "Pakistan",
    "Nigeria", "Brazil", "Bangladesh", "Russia", "Ethiopia"
]
populations = [
    1450935791, 1419321278, 345426571, 283487931, 251269164,
    232679478, 211998573, 173562364, 144820423, 132059767
]

sum_populations = sum(populations)
print(f"The sum of the top ten population is: {sum_populations}")

percentages = []

for population in populations:
    times_pop = population * 100
    percent = times_pop / sum_populations
    
    percentages.append(f"{percent.__round__(2)}%")

print(f"Countries population base on percentages: {percentages}")
```

```python
xpoints = np.array(percentages)
ypoints = np.array(countries)

plt.title("Top 10 countries in the world by population (2024)")
plt.ylabel("Top 10 countries")
plt.xlabel("Countries populations")

plt.bar(xpoints, ypoints, color="hotpink")
```

```python
nums = np.array([i for i in range(1, 11)])

index_num = "0123456789"
reverse_index_num = index_num[::-1]

items = np.array([i for i in reverse_index_num])

plt.pie(
    nums,
    labels = items,
    explode = [i/100 for i in range(1, 11)],
    colors = [
        "r", "g", "b", "grey", "aqua",
        "m", "c", "y", "b", "dodgerblue"
    ]
)

plt.title("Survey with pie-chat", color="green")

states = [f"{idx}, {c}" for idx, c in enumerate(countries)]
plt.legend(
    title = "Countries base on ranks:",
    loc = "upper left",
    labels = states
)
```
