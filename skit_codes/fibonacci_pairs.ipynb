```python
# Fibonacci pairs that asks how many
# fib numbers you would like to see and
# then returns that many numbers from the sequence,
# lastly visualise it with matplotlib library.
```

```python
fib = [0, 1]
n = int(input())

for i in range(2,n):
  fib.append(fib[i-1] + fib[i-2])

pairs = []

for i in range(n-1):
  pairs.append([fib[i], fib[i+1]])

print(pairs)
```

```python
import matplotlib.pyplot as plt
import numpy as np

pairs = np.array(pairs)

plt.scatter(pairs[:, 0], pairs[:, 1])
plt.grid()
plt.show()
```

```python
plt.plot(pairs[:, 0], pairs[:, 1], 'r:*')
plt.grid()
```

```python
plt.bar(pairs[:, 0], pairs[:, 1])
plt.grid()
```
