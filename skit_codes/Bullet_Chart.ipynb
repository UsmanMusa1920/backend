```python
import matplotlib.pyplot as plt

categories = ["Category"]
ranges = [(50, 100)]

fig, ax = plt.subplots()
ax.barh(categories, [75], color="lightblue")

for i, (low, high) in enumerate(ranges):
    ax.plot([low, high], [i] * 2, color="black")
    ax.plot([85], [0], marker="o", markersize=10, color="red")

plt.title("Bullet Chart")
plt.show()
```
